```python
import pandas as pd
import docx2txt
import glob

# Create empty lists to store data and file names
data = []
file_names = []

# Use glob to get paths of all .doc and .docx files in folder
files = glob.glob("C:\\Users\\rajiv\\OneDrive\\Desktop\\resumedataset\\Resumes\\Peoplesoft resumes\\*.docx") + glob.glob("C:\\Users\\rajiv\\OneDrive\\Desktop\\resumedataset\\Resumes\\Peoplesoft resumes\\*.doc")

# Loop through files
for file in files:
    # Get file name and append to file_names list
    file_name = file.split("\\")[-1]
    file_names.append(file_name)
    # Read text data from file and append to data list
    if file_name.endswith('.docx'):
        text = docx2txt.process(file)
    else:
        with open(file, 'rb') as f:
            text = str(f.read())
    data.append(text)

# Create a dataframe
df = pd.DataFrame({'File Name': file_names, 'Text Data': data})
```
